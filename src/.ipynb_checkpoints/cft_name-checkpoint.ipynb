```python
import pandas as pd
```

```python
train_df = pd.read_csv('../data/train.csv')
```

```python
train_df.shape
```

```python
cns = train_df.country.value_counts()
top_countries = list(cns[cns>200].index)
train_df['country'] = train_df['country'].apply(lambda x: x if x in top_countries else 'OTHER')
```

```python
train_df.country.value_counts()
```

```python
pd.pivot_table(train_df,
               index='country',
               columns='target',
               values='fullname',
               aggfunc=pd.Series.nunique)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase=False,
                     ngram_range=(1,1),
                     analyzer='char')
```

```python
cv.fit(train_df.fullname)
```

```python
rare_char = "'"
train_df[train_df.fullname.apply(lambda x: True if rare_char in x else False)]
```

```python
rare_char = 'Ї'
train_df[train_df.fullname.apply(lambda x: True if rare_char in x else False)]
```

```python
train_df[train_df.fullname.apply(lambda x: True if 'а' in x else False)]
```
